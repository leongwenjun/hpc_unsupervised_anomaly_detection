```python
import pandas as pd

def reshape_df(df, metric_name):
    df = df.stack().reset_index()
    df.columns = ['time', 'node', metric_name]
    return df

node_load15_df = pd.read_parquet(node_load15, engine='pyarrow')
node_load1_df = pd.read_parquet(node_load1, engine='pyarrow')
node_load5_df = pd.read_parquet(node_load5, engine='pyarrow')
memory_df = pd.read_parquet(memory, engine='pyarrow')
gpu_temp_df = pd.read_parquet(gpu_temp, engine='pyarrow')
gpu_power_df = pd.read_parquet("./encryptedParq/pqNvidiaEncrypted/nvidia_gpu_power_usage_milliwatts", engine='pyarrow')

node_load15_df = reshape_df(node_load15_df, 'load15')
node_load1_df = reshape_df(node_load1_df, 'load1')
node_load5_df = reshape_df(node_load5_df, 'load5')
memory_df = reshape_df(memory_df, 'memory_active')
gpu_temp_df = reshape_df(gpu_temp_df, 'gpu_temp')
gpu_power_df = reshape_df(gpu_power_df, 'gpu_power')

merged_df = node_load15_df

for df in [node_load1_df, node_load5_df, memory_df, gpu_temp_df, gpu_power_df]:
    if df is gpu_temp_df or df is gpu_power_df:
        merged_df = merged_df.merge(df, on=['time', 'node'], how='outer').fillna(0)
    else:
        merged_df = merged_df.merge(df, on=['time', 'node'], how='outer').dropna()

print(merged_df.head())
```

```python
mapping_df = pd.read_csv('decreptedNode.csv', header=None, names=['encrypted', 'decrypted'])
mapping_dict = dict(zip(mapping_df['encrypted'], mapping_df['decrypted']))

merged_df['node'] = merged_df['node'].map(mapping_dict)

print(merged_df.head())
```

```python
column_mapping = {
    'time': 'timestamp',
    'node': 'node',
    'load15': 'node_load15',
    'load1': 'node_load1',
    'load5': 'node_load5',
    'memory_active': 'node_memory_Active_bytes',
    'gpu_temp': 'nvidia_gpu_temperature_celsius-max',
    'gpu_power': 'nvidia_gpu_power_usage_milliwatts-max'
}

merged_df = merged_df.rename(columns=column_mapping)
```
